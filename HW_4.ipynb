### Условие 3: 3 задача
#### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
#### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
# 1. График зависимости стоимости от квадратуры жилой площади 
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sqft_living', y='price', alpha=0.7)
plt.xlabel('Квадратура жилой площади ')
plt.ylabel('Стоимость недвижимости ')
plt.title('Зависимость стоимости от квадратуры жилой площади')
;
```

##### Из графика видно, что есть зависимость между квадратурой жилой площади и стоимостью недвижимости - чем больше площадь дома, тем выше его стоимость.

```python
# 2. Зависимость стоимости недвижимости от количества комнат 
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='bedrooms', y='price')
plt.xlabel('Количество комнат')
plt.ylabel('Стоимость недвижимости')
plt.title('Стоимости недвижимости по количеству комнат')
;
```

##### Дома с большим количеством комнат могут иметь более высокую стоимость, но есть также выбросы с низкой стоимостью для некоторых комбинаций количества комнат.

```python
# 3. Зависимость стоимости недвижимости от вида на набережную
sns.boxplot(x=df['waterfront'],y=df['price'])
plt.title('Стоимость от вида на набережную')
plt.ylabel('Стоимость недвижимости')
plt.xlabel('Вид на набережную')
;
```

##### На графике видим прямую зависимость от вида на набережную. Цена сильно зависит от вида

```python
# 4. Зависимость стоимости недвижимости от года постройки
sns.boxplot(x=df['view'],y=df['price'])
plt.title('Стоимость от от года постройки')
plt.ylabel('Стоимость недвижимости')
plt.xlabel('Год постройки')
;
```

##### здесь видим прямую зависимость от вида, чем лучше вид, тем выше цена
